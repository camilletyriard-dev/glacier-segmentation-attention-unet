# 2. Identify a local challenge

**Country**: Switzerland

**Region**: Swiss Alps Jungfrau–Aletsch area, focusing on the Great Aletsch Glacier and neighbouring glacierised catchments.

**Challenge**: With global warming, glaciers, among others, in the Swiss Alps, are melting. The Attention U-Net application on deforestation could be relevant here to transfer its knowledge to another application: segmentation of ice/non-ice instead of forest/non-forest. This could help in continuous, high-resolution monitoring and segmentation of glacier extent in the Jungfrau–Aletsch region to track rapid retreat, quantify area change, and detect emerging high-risk zones (e.g. newly formed pro-glacial lakes or exposed unstable slopes) during the summer months.

**Problems**: 
- With climate change and global warming, glaciers are shrinking.  Glacier-evolution modelling for Aletsch shows that 60 % of today’s ice volume could be lost by 2100 for a moderate emission of CO2, and an almost complete loss is expected under the most extreme emission scenario. [1]
- The Swiss Alps have lost about a quarter of their volume since 2015, with 3 % loss in 2025 alone, the 4th-largest retreat on record. Over 1,000 small glaciers have already disappeared. [2]
- Since 2015, one quarter of the Swiss glacier volume has been lost. [4]

**Effects/Why important**:
- **Water resources**:  
The Alps function as one of Europe’s largest freshwater reservoirs, supplying an estimated 170 million people across major river basins such as the Rhine, Rhone, Po and Danube. Stable access to Alpine water is essential for drinking water provision, agriculture, industry and hydropower generation. Yet climate change is disrupting these systems. Glacier retreat already reduce the reliability of seasonal water flows, and future shortages are expected to become a critical challenge in the coming decades. [8]

- **Hydropower & Agriculture**:
Meltwater from Alpine glaciers feeds key Swiss river systems that are central to hydropower production and irrigation. As glaciers shrink, the timing and availability of meltwater shift. In the short term, hydropower may experience increased flows during accelerated melting, but once “peak meltwater” is passed, long-term water availability declines. This means hydropower output becomes more volatile and agricultural regions face stronger competition for water during prolonged heat events, when demand is already at its highest.

- **Natural hazards**:
Glacier retreat reshapes mountain landscapes in ways that increase safety risks. Thinning ice exposes unstable slopes and encourages the formation of new glacial lakes, which heightens the likelihood of rockfalls, ice avalanches and glacial lake outburst floods. These hazards pose direct risks to mountain communities and infrastructure such as hiking routes, ski lifts and high-altitude huts. A recent example is the May 2025 glacier collapse near Blatten, Switzerland, where large parts of the village were buried and destroyed, as seen in the image below. [5]

The consequences of glacier loss extend far beyond the Alps. As ice melts, global sea levels rise, increasing the vulnerability of coastal cities well outside Switzerland. At the same time, warmer oceans and shifting currents disrupt global weather patterns, contributing to more frequent storms, heavier flooding and greater climate instability. These changes directly affect marine ecosystems: warmer waters alter where and when fish spawn, placing pressure on fisheries and the communities whose livelihoods depend on them. Coastal regions also face rising economic costs as extreme weather events and flooding become more common, resulting in repeated and increasingly expensive recovery efforts. [6,7]

- **Biodiversity**:
Glaciers and their surrounding environments support specialised species that are uniquely adapted to extreme cold. As ice continues to disappear, these ecosystems face increasing pressure, with many species at heightened risk of extinction. This challenge is most visible in regions such as Greenland and Antarctica, where polar species, including the polar bear, are already experiencing severe habitat loss. Alpine biodiversity, however, is also vulnerable, as many cold-adapted plants and animals depend on stable glacial conditions to survive. Similar patterns are seen across the Arctic. The rapid decline of sea ice is forcing species such as walrus and polar bears to abandon traditional habitats. Polar bears are spending more time on land in search of food, which leads to more frequent encounters and conflict with human communities [6,7]


- **Tourism, culturale value and global heritage**:
The Aletsch Glacier is both an iconic symbol of the Swiss Alps and a central attraction for glacier hikes, panoramic viewpoints and educational tours. Its retreat reduces the visual and cultural value of the landscape and makes certain routes more unstable or unsafe. Ski seasons shorten, guided glacier visits become more restricted, and local economies, particularly those relying on skiing and nature-based tourism, experience increasing uncertainty and need to be innovative to adapt to the changing landscape. Lower-altitude ski areas already struggle with shorter seasons and unreliable snow conditions. In addition, the Aletsch Glacier forms the heart of the Jungfrau–Aletsch UNESCO World Heritage Site. Rapid glacier loss threatens this status and could result in a decline in global natural heritage.

**Can this AI method realistically help?**

Using an Attention U-Net for glacier segmentation can support climate adaptation in alpine regions, but only under specific technical and institutional conditions.


Satellite imagery is well suited for glacier monitoring, and Attention U-Net can be used for accurate and quick segmentation tasks. It can detect glacier boundaries, proglacial lakes and surface changes more frequently and at lower cost than field-based monitoring.  
However, seasonal snow, shadows, debris cover and cloud contamination can significantly reduce model reliability as the model can misclassify it as ice.

When integrated into existing monitoring frameworks, the model can enhance hazard detection (e.g., glacial lake growth, slope instability) and improve long-term planning for water management and hydropower. The model can serve as an extra opninion, but not to replace experts. Experts, someone that knows the mountains, remain invaluable in evaluating the conditions.


However, as with all implementations, these AI model insights only create real impact if embedded in institutions that can act on them. There is a need for governance, ai engineers, and funding to determine wheter this information can lead to valuable operational decisions.

**Scalability**

- Sentinel-2 data is globally available, allowing the model to scale beyond Switzerland. However, domain shift (e.g., debris cover, snow variability) may require fine-tuning for each region.
- The architecture is compact and can be trained on modest GPU resources, making it accessible for research institutes and national environmental agencies.
- Once trained, the model can process new imagery efficiently, enabling near-real-time glacier monitoring and supporting hazard early-warning systems.


**Limitations and Ethical Considerations**

- **Seasonal variability:** Snow cover, cloud shadows and debris-covered ice can reduce model reliability. Bright clouds may be misclassified as glacier ice, and debris-covered sections are often difficult for the model to detect.
- **Sparse labelled data:** High-quality pixel-level annotations for glaciers are limited and expensive to produce. This limits the model’s ability to generalise across seasons or different glacier types.
- **Dependence on satellite quality:** Limited cloud-free imagery constrain model performance, especially outside summer months.

- **Risk of overreliance on automated outputs:** Authorities may overtrust the model without proper uncertainty estimation. In hazard forecasting (e.g., glacial lake outburst floods), false negatives can have severe consequences.
- **Transparency and interpretability:** Deep-learning segmentation models provide limited interpretability. Other methods need to be in place to increase trust and explain model predictions.
- **Data governance and accessibility:** While raw satellite data is openly available, access to high-quality processed glacier datasets may have licensing restrictions. For this application, we labelled and processed the data ourselves.
- **Environmental impact of computation:** Although Attention U-Net is considerably more efficient than many larger deep-learning models, its training and maintenance still consume energy. The source of the energy used to train and deploy these models becomes equally important. Training on renewable-powered infrastructures reduces the overall climate impact, whereas fossil-fuel-based compute can undermine the environmental benefits the model aims to support.

- **Social implications:** Glacier retreat insights may affect tourism revenues, insurance risk assessments and municipal planning. Transparent communication is essential to ensure that communities, policymakers and businesses have equal access to information and are not disadvantaged by unequal data availability.




 **Overall evaluation:**
The model is **scalable, low-cost and environmentally responsible** when deployed with renewable-powered compute and strong institutional governance. Its long-term sustainability depends on transparent documentation, expert validation and responsible communication of uncertainties.
